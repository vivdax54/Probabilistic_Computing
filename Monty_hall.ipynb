```python
import numpy as np
import matplotlib.pyplot as plt

doors = [1, 2, 3]
num_of_experiments = 10000

# the car is always behind door 1
success_if_changed = np.zeros(num_of_experiments)
success_if_not_changed = np.zeros(num_of_experiments)

for i in range(num_of_experiments):

    # Step 1: initial choice
    first_choice = np.random.choice(doors)

    # Step 2: host opens a goat door
    if first_choice == 1:
        opened = np.random.choice([2, 3])
    else:
        opened = 3 if first_choice == 2 else 2

    # Step 3: switching choice
    second_choice = [d for d in doors if d not in [first_choice, opened]][0]

    # Step 4: evaluate outcomes
    # switching success
    if second_choice == 1:
        success_if_changed[i] = 1

    # not switching success
    if first_choice == 1:
        success_if_not_changed[i] = 1



cum_success_switch = np.cumsum(success_if_changed) / np.arange(1, num_of_experiments + 1)
cum_success_stay = np.cumsum(success_if_not_changed) / np.arange(1, num_of_experiments + 1)



plt.figure(figsize=(6, 4))
plt.plot(cum_success_switch, label="Switch", linestyle='--')
plt.plot(cum_success_stay, label="Stay")

# theoretical lines
plt.axhline(2/3, linestyle=':', label="2/3 (Switch)")
plt.axhline(1/3, linestyle=':', label="1/3 (Stay)")

plt.xlabel("Number of Experiments")
plt.ylabel("Success Probability")
plt.title("Monte Carlo Convergence (Monty Hall)")
plt.legend()
plt.grid(True)

plt.show()
```
