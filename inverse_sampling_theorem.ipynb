```python
######################## Applying Inverse Transform Sampling ########################

import numpy as np
import matplotlib.pyplot as plt

#### sampling from uniform distribution
num_samples  = 100000
uniform_samples = np.random.uniform(0, 1, num_samples)


### apply Inverse Sampling Theorem
logistic_samples = np.log(uniform_samples/(1-uniform_samples))

#### analytical inverse sampling theorem
def logist(x, loc=0, scale=1):
    return np.exp((loc-x)/scale)/(scale*(1+np.exp((loc-x)/scale))**2)


plt.figure(figsize=[10, 4])
plt.suptitle('Inverse Transform Sampling', fontsize=14)
plt.subplot(1, 2, 1)
count, bins, ignored = plt.hist(uniform_samples, 100, density=True, alpha=0.75, label='Simulation')
plt.plot(bins, np.ones_like(bins), linewidth=2, color='r', label='Analytical')
plt.title('Uniform distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()

plt.subplot(1, 2, 2)
count, bins, ignored = plt.hist(logistic_samples, 100, density=True, alpha=0.75, label='Simulation')
lgst_val = logist(bins)
plt.plot(bins, lgst_val, color='r', label='Analytical')
plt.title('Logistic distribution')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```
