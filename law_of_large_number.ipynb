```python
#### Toy example to Illustrate the Law of Large number #####

import numpy as np
import matplotlib.pyplot as plt

three_sequences_coin_tosses = np.random.choice([0,1], size=(3, 1000))

# one way to compute the time average is to use cumulative sum and divide by the number of tosses at each step
time_avg = np.cumsum(three_sequences_coin_tosses, axis=1) / np.arange(1, 1001)



plt.figure(figsize=(7,6))
plt.plot(np.linspace(1, 1000,1000), time_avg[0], alpha=0.5, label='Sequence 1')
plt.plot(np.linspace(1, 1000,1000), time_avg[1], alpha=0.5, label='Sequence 2')
plt.plot(np.linspace(1, 1000,1000), time_avg[2], alpha=0.5, label='Sequence 3')
plt.axhline(0.5, color='red', linestyle='--', label='True Probability (p=0.5)')
plt.xlabel('Number of Tosses')
plt.ylabel('Time Average (Proportion of Heads)')
plt.title('Law of Large Numbers: Coin Toss Outcomes')
plt.legend()
plt.grid()
plt.show()

```
